# Importing Libraries

```python
import numpy as np
import pandas as pd
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import lightgbm as lgb
from tqdm import tqdm
```

# Loading Data

```python
ntr = 7000
nts = 3000

train1 = pd.read_csv('train.csv',nrows=ntr)
names = ['msno','song_id','target','source_system_tab','source_screen_name','source_type']
```

```python
songs = pd.read_csv('songs.csv')
members = pd.read_csv('members.csv')
```

### Arranging Data of Song Dataset

```python
songs.columns
```

```python
train1.columns
```

```python
print("Data Processing.....")
song_cols = ['song_id','artist_name','genre_ids','song_length','language']
train1 = train1.merge(songs[song_cols],on='song_id',how='left')
test1 = test1.merge(songs[song_cols],on='song_id',how='left')
```

# Splitting into year,month and date

```python
members.head()
```

```python
members['registration_year'] = members['registration_init_time'].apply(lambda x: int(str(x)[0:4]))
members['registration_month'] = members['registration_init_time'].apply(lambda x: int(str(x)[4:6]))
members['registration_date'] = members['registration_init_time'].apply(lambda x: int(str(x)[6:8]))
```

```python
members.head()
```

### So we got seperate Date and Time Dropping Registration Time Column

```python
members = members.drop(['registration_init_time'],axis=1)
```

```python
members.head()
```

# Left join with trining a testing data

```python
members_cols = members.columns
train1 = train1.merge(members[members_cols],on='msno',how='left')
test1 = test1.merge(members[members_cols],on='msno',how='left')
train1 = train1.fillna(-1)
test1 = train1.fillna(-1)
train1.head()
```

# Collection work for Members and songs 

```python
import gc
del members, songs; gc.collect()
```

```python
cols = list(train1.columns)
cols.remove('target')
train1.head()
```

```python
test1.head()
```

# Label Encoder

```python
for col in tqdm(cols):
    if train1[col].dtype == 'object':
        train1[col] = train1[col].apply(str)
        test1[col] = test1[col].apply(str)
        
        le = LabelEncoder()
        train1_vals =list(train1[col].unique())
        test1_vals = list(test1[col].unique())
        le.fit(train1_vals + test1_vals)
        train1[col] = le.transform(train1[col])
        test1[col] = le.transform(test1[col])
```

# Song Popularity

```python
unique_songs = range(max(train1['song_id'].max(), test1['song_id'].max()))
song_popularity = pd.DataFrame({'song_id': unique_songs, 'popularity':0})

train_sorted = train1.sort_values('song_id')
train_sorted.reset_index(drop= True, inplace=True)
test1_sorted = test1.sort_values('song_id')
test1_sorted.reset_index(drop=True, inplace=True)
```

# Model Training and prediction

```python
# User library size
x = np.array(train1.drop(['target'], axis=1))
y = train1['target'].values

#x_test = np.array(test1.drop(['id'], axis=1))
#ids = test1['id'].values

del train1, test1; gc.collect();

x_train, x_valid, y_train, y_valid = train_test_split(x, y,test_size=0.1, random_state = 12)
    
del x, y; gc.collect();

d_train = lgb.Dataset(x_train, label=y_train)
d_valid = lgb.Dataset(x_valid, label=y_valid) 

watchlist = [d_train, d_valid]
```

```python
print("Training LGBM model....")
params = {}

params['learning_rate'] = 0.4
params['application'] = 'binary'
params['max_depth'] = 15
params['num_leaves'] = 2**8
params['verbosity'] = 0
params['emtric'] = 'auc'

model = lgb.train(params, train_set=d_train, num_boost_round=200, valid_sets=watchlist,early_stopping_rounds=10, verbose_eval=10)
```

```python
print("Making prediction and saving them.....")
p_test =  model.predict(x_train)
p_test
```

# Creating csv file to store ids and  traget columns

```python
sub = pd.DataFrame()
#sub['id'] = ids
sub['target'] = p_test
sub.to_csv('submission.csv.gz', compression='gzip', index=False, float_format='%.5f')
print('Done')
```

```python
yhat = (p_test>0.5)
comp = (yhat==ytr) 
acc = comp.sum()//comp.size*100
print('The accuracy of lgbm model on test data is: {0:f}%'.format(acc))
```
